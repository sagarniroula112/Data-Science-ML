# Lets perform Encoding

## Applying One-Hot Encoding

<h3> First Do it with pandas </h3>

```python
# Get dummies with pandas in any variable
country_dummy = pd.get_dummies(dataset['country'], dtype='int')
country_dummy
```

```python
# concatenate the dataframes into original dataframes.
dataset = pd.concat([dataset, country_dummy], axis = 1)
dataset
```

```python
# Drop the country column and rearrange the index of column.
dataset.drop('country', axis = 1, inplace=True)
dataset.head(10)
```

```python
dataset_1 = dataset[['Australia', 'Canada', 'Dubai', 'USA', 'Salary', 'YearsExperience', 'Purchased']]
dataset_1.head(20)
```

# <h3> Let's do it with Scikit-Learn </h3>
Also Restart your kernel if you are using same notebook


```python
# Lets first perform label encoding
# Labelencoding will be performed on Purchased column as it has only two unique value in it.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset_2 = pd.read_csv("Salary_Dataset.csv")

dataset_2['Purchased'] = le.fit_transform(dataset_2['Purchased'])
dataset_2.head()
```

```python
# Lets perform one-hot encoding on country column.
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])] , remainder= 'passthrough')
dataset_2 = pd.DataFrame(ct.fit_transform(dataset_2))

dataset_2
```

```python
#Lets link the headers name.
dataset_2.columns = ['Australia', 'Canada', 'Dubai', 'USA', 'Salary', 'YearsExperience', 'Purchased']
dataset_2
```
